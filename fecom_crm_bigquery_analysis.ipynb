Bigquery Setup

```python
from google.cloud import bigquery
import pandas as pd
import numpy as pd
import os 

#This project initializes the BigQuery client to establish a connection between the Python application and
# the Google BigQuery
client = bigquery.Client(project='sql-sandbox-435415')
```

---

Let's preview the first few rows of each core table used in this analysis:
- customer_list
- orders
- order_items
- sellers_list

```python
sql_query ="""
SELECT 
*
FROM 
`sql-sandbox-435415.fecom_crm.customer_list`
LIMIT 10
"""

query_job = client.query(sql_query)
df_customer_list = query_job.to_dataframe()
df_customer_list
```
